#Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Load the Dataset

```python
df = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv")
```

```python
df.head()
```

```python
df.info()
```

#Check Columns

```python
df.columns
```

#Bar Chart (Categorical Variable – Region)

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df, x="Region")

plt.xticks(rotation=45)
plt.title("Distribution of Countries by Region")
plt.xlabel("Region")
plt.ylabel("Number of Countries")

plt.show()
```

#Bar Chart (Income Group Distribution)

```python
plt.figure(figsize=(10,6))

sns.countplot(data=df, x="IncomeGroup")

plt.xticks(rotation=45)
plt.title("Distribution of Countries by Income Group")
plt.xlabel("Income Group")
plt.ylabel("Count")

plt.show()
```
